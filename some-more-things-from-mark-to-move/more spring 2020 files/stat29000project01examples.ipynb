```python
import csv
```

```python
# Let's read in a dataset to practice with
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# What if we wanted just the header row?
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)
    print(header)
```

```python
# Lets access just the year column
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    for youCanNameThisVariableAnything in reader:
        print(youCanNameThisVariableAnything[3])
```

```python
# let's print the entire row, but only if year is > 2008
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    
    for row in reader:
        if row[3] > 2008:
            print(row)
            
# oh no! an error!
```

```python
# the error tells us row[3] is a 'str' and 2008 is an 'int'
# lets convert the 'str' to an 'int'
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    
    for row in reader:
        if int(row[3]) > 2008:
            print(row)
```

```python
majors = []
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    
    for row in reader:
        majors.append(row[1])

# we can use sets to get a unique list of "major_field" very very quickly
result = set(majors)
print(result)

# we can also get a count of each major using dicts
counts = dict()
for major in majors:
    counts[major] += 1
    
# oh no! another error!
```

```python
# we got a "KeyError" here because for a dict() we must
# define all of the keys ahead of time. We can get around
# this...
from collections import defaultdict

counts = defaultdict(int)
for major in majors:
    counts[major] +=1
    
print(counts)
```

```python
# thats not very appealing...
for key, value in counts.items():
    print(f'{key}: {value}')
```

```python
# lets experiment with list comprehensions
majors = [row for row in majors if "science" in row]
print(set(majors))
```

```python
# Let's write out a subset of the data to a new file
# But lets make it a tab spaced instead of comma
with open("phd_by_field.csv", "r") as file:
    reader = csv.reader(file)
    with open("new_file.tsv", "w") as file_to_write:
        writer = csv.writer(file_to_write, delimiter='\t')
        for row in reader:
            writer.writerow(row)
```
